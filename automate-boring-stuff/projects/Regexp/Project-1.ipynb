```python
import pyperclip, re
phone_pattern = re.compile (r'''(
                            (\+?\d{2,5}|\(\+\d{2,5}\))? # Area Code - Optional
                            (\s|-|\.)? # Seperator - Optional
                            (\d{4}) # First four digits
                            (\s|-|\.)? # Seperator - Optional
                            (\d{2,6}) # Last digits
                            (\s+(ext|X|ext\.)(\s|-)?(\d{3,5}))? # Extension
                          )''', re.VERBOSE)
text = str(pyperclip.paste())
ph_output = []
if not text:
    print("Clipboard is empty! Copy some text")
else:
    matches = phone_pattern.findall(text)
    if not matches:
        print ("No phone number found in the copied text!")
    else:
        for match in matches:
            ph_output.append(match[0])

if len(ph_output) > 0:
    pyperclip.copy('\n'.join(ph_output))
    print('Phone numbers copied to clipboard')
else:
     print("No phone numbers found")
```

```python
import pyperclip, re
# email pattern
email_pattern = re.compile (r'''(
                            ([a-zA-Z0-9._%+-]+) # User name
                            @ #                 @ symbol
                            ([a-zA-Z0-9_-]+) # Domain name
                            (\.)             # Seperator
                            ([a-zA-Z]{2,4}) # 1st domain name last qualifier
                            (\.)?           # Seperator - optional
                            ([a-zA-Z]{2,4})? # 2nd domain name last qualifier - optional
                            )''', re.VERBOSE)
# phone pattern
phone_pattern = re.compile (r'''(
                            (\+?\d{2,5}|\(\+\d{2,5}\))? # Area Code - Optional
                            (\s|-|\.)? # Seperator - Optional
                            (\d{4}) # First four digits
                            (\s|-|\.)? # Seperator - Optional
                            (\d{2,6}) # Last digits
                            (\s+(ext|X|ext\.)(\s|-)?(\d{3,5}))? # Extension
                          )''', re.VERBOSE)
consolidated_match = []
# Find matches in clipboard text.
text = str(pyperclip.paste())
# search phone and email addresses
if not text:
    print("Clipboard is empty! Copy some text")
else:
    matches = email_pattern.findall(text) # email search
    if not matches:
        print ("No email found in the copied text!")
    else:
        for match in matches:
            consolidated_match.append(match[0])
    matches = phone_pattern.findall(text)
    if not matches:
        print ("No phone number found in the copied text!")
    else:
        for match in matches:
            consolidated_match.append(match[0])
if len(consolidated_match) > 0:            
    # consolidated_match.join(msg)
    pyperclip.copy('\n'.join(consolidated_match))
    print('Copied to clipboard')
else:
    print('No email address and/or phone number found in copied text')
```
