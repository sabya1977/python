```python
import pyperclip, re
url_pattern = re.compile(r'''(
                         (?:http|https)
                         :
                         /{2}
                         (?:www\.)?
                         (?:[a-zA-Z0-9.@_:+=*#~%()$`&?!-/]+)
                         )''', re.VERBOSE)
consolidated_match = []
# Find matches in clipboard text.
text = pyperclip.paste()
# search for urls
if not text:
    print("Clipboard is empty! Copy some text")
else:
    matches = url_pattern.findall(text)                 # urls search
    if not matches:
        print ("No urls found in the copied text!")
    else:
        for match in matches:
            # print(match)
            consolidated_match.append(match)
# print(consolidated_match)

if len(consolidated_match) > 0:            
    pyperclip.copy('\n'.join(consolidated_match))
    print('Copied to clipboard')
else:
    print('No urls found in copied text')
```

```python
import pyperclip, re
url_pattern = re.compile(r'(https?://(www\.)?([a-zA-Z0-9.@_:+=*&?!-/]+))')
text = 'Check out my channel: https://www.youtube.com/@user-123 and my site http://example.com'
matches = re.finditer(url_pattern, text)
for match in matches:
    print(match.group(0))
    print(match.start())
    print(match.end())
```
