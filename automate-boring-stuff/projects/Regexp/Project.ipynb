```python
import pyperclip, re
email_pattern = re.compile (r'''(
                            ([a-zA-Z0-9._%+-]+) # User name
                            @ #                 @ symbol
                            ([a-zA-Z0-9_-]+) # Domain name
                            (\.)             # Seperator
                            ([a-zA-Z]{2,4}) # 1st domain name last qualifier
                            (\.)?           # Seperator - optional
                            ([a-zA-Z]{2,4})? # 2nd domain name last qualifier - optional
                            )''', re.VERBOSE)
text = str(pyperclip.paste())
output = []
if not text:
    print("Clipboard is empty! Copy some text")
else:
    matches = email_pattern.findall(text)
    if not matches:
        print ("No email found in the copied text!")
    else:
        for match in matches:
            output.append(match[0])

if len(output) > 0:
    pyperclip.copy('\n'.join(output)) # takes elements from output list and make a string seperated by \n
    print('Copied to clipboard')
    print('\n'.join(output)) # takes elements from output list and make a string seperated by newline
else:
    print("No email address found")
```
